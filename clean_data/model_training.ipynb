```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
data_file='DATA.pkl'
data=pd.read_pickle('DATA.pkl')
data
```

# alpha = symmetric

```python
from sklearn.metrics.cluster import adjusted_rand_score

def randIndexMetric(predictions,data):
    pred_topic=[]
    for pred in predictions:
        topic_prob=sorted(pred,key=lambda x:-x[1])
#         print(topic_prob)
        top_topic=topic_prob[0][0]
        pred_topic.append(top_topic)
    _data=data.copy()
    _data['pred_topic']=pred_topic
    ari=adjusted_rand_score(_data['module'], _data['pred_topic'])
#     print(ari)
    return ari
```

```python
%%time
from gensim.corpora import Dictionary
from gensim.models import LdaModel,LdaMulticore

docs=data['motif_list'].values
dictionary=Dictionary(docs)
corpus = [dictionary.doc2bow(doc) for doc in docs]

temp = dictionary[0]
ari_list=[]
num_runs=10
for i in range(num_runs):
    print('Run No. = ',i)
    lda = LdaModel(corpus, 
                   id2word=dictionary.id2token,
                   alpha='symmetric',
                   chunksize=10000,
                   eta='auto',
                   num_topics=3, 
                   iterations=1000, 
                   passes = 10,
                   minimum_probability=0.0
                  )
    print('Alpha = ',lda.alpha)
    predictions=lda.get_document_topics(corpus,minimum_probability=0.0)
    ari=randIndexMetric(predictions,data)
    ari_list.append(ari)
print()
print('All ARIs:',ari_list,'\n')
print(f'Average ARI ({num_runs}) =',np.mean(ari_list),'\n')
```
