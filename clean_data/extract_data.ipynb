```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
data_size=1000
index=[np.arange(450),np.arange(450,900),np.arange(900,1000)]
```

```python
df1=pd.DataFrame()
df1['seq_no.']=index[0]
df1['module']=1
df1['motif_1']=1
df1['motif_2']=1
df1['motif_3']=0
df1['motif_4']=0
df1['motif_5']=0
df1['motif_6']=0
df1
```

```python
df2=pd.DataFrame()
df2['seq_no.']=index[1]
df2['module']=2
df2['motif_1']=0
df2['motif_2']=0
df2['motif_3']=1
df2['motif_4']=1
df2['motif_5']=0
df2['motif_6']=0
df2
```

```python
df3=pd.DataFrame()
df3['seq_no.']=index[2]
df3['module']=3
df3['motif_1']=0
df3['motif_2']=0
df3['motif_3']=0
df3['motif_4']=0
df3['motif_5']=1
df3['motif_6']=1
df3
```

```python
data=pd.concat([df1,df2,df3],ignore_index=True)
data
```

```python
data['module'].value_counts()
```

```python
def process_row(row):
    motif_list=[]
    if row['motif_1']==1:
        motif_list.append('motif_1')
    if row['motif_2']==1:
        motif_list.append('motif_2')
    if row['motif_3']==1:
        motif_list.append('motif_3')
    if row['motif_4']==1:
        motif_list.append('motif_4')
    if row['motif_5']==1:
        motif_list.append('motif_5')
    if row['motif_6']==1:
        motif_list.append('motif_6')
    if len(motif_list)==0:
        motif_list.append('none')
    return ','.join(motif_list)
data['motif_string']=data.apply(process_row,axis=1)
data['motif_list']=data['motif_string'].apply(lambda x:x.split(','))
data
```

```python
data.to_csv('DATA.csv',index=False)
data.to_pickle('DATA.pkl')
```
